### Import Thư Viện

```python
import pandas as pd
import numpy as np
import requests as r
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

### 1. Đọc dữ liệu từ file CSV đã được xử lí

```python
new_track = pd.read_csv('New_Track.csv')
new_track
```

```python
new_track['id'] = new_track['id'].astype('string')
```

```python
new_track.info()
```

#### Tóm tắt thông tin về 9 thuộc tính của tập dữ liệu Track:

|Thuộc tính|Kiểu dữ liệu|     Ý nghĩa  
|:--------:|:-------------:|:-----------:
|**id**|Số nguyên|Định danh cho mỗi track|
|**full_duration**|Số nguyên|Thời gian của từng track|   
|**genre**|Chuỗi|Thể loại nhạc của mỗi track| 
|**comment_count**|Số nguyên|Số lượt bình luận track đó|
|**likes_count**|Số nguyên|Số lượt thích track đó|
|**playback_count**|Số nguyên|Số lượt nghe lại track đó|   
|**download_count**|Số nguyên|Số lượt tải về của track đó| 
|**display_date**|Chuỗi|Thời gian xuất bản track lên Soundcloud|

### 2. Phân tích dữ liệu đơn biến

```python
track_num = new_track.select_dtypes(include = ['float64', 'int64'])
track_num.describe().T                                
```

#### Biểu đồ tuần suát của các biến numeric

```python
track_num.hist(figsize=(16, 16), bins=150, xlabelsize=8, ylabelsize=8);
```

#### Nhận xét sơ bộ: 
    + full_duration: Các track đa số kéo dài khoảng 5 - 10 phút
    + comment_count, reposts_count và download_count: Rất ít đa số track đều rất ít có comment, repost và download.
    + likes_count: Có só lượt yêu thích trong khoảng từ 0 đến 2000 
    + playback_count: Có số lượt playback trong khoảng từ 0 đến 50000

```python
figure, axis = plt.subplots(2, 3, figsize=(15,15))

axis[0][0].boxplot(new_track["full_duration"])
axis[0][0].set_title("full_duration")

axis[0][1].boxplot(new_track["likes_count"])
axis[0][1].set_title("likes_count")

axis[0][2].boxplot(new_track["reposts_count"])
axis[0][2].set_title("reposts_count")

axis[1][0].boxplot(new_track["download_count"])
axis[1][0].set_title("download_count")

axis[1][1].boxplot(new_track["comment_count"])
axis[1][1].set_title("comment_count")

axis[1][2].boxplot(new_track["playback_count"])
axis[1][2].set_title("playback_count")

plt.show()
```

Nhận xét

+ full_duration: Đa số các track đề có thời lượng trong khoảng 5-15 phút và có 1 track đến 25 gần phút
+ likes_count: Đa số lượt like của track trong khoảng từ 0-2000 và có xuất hiện một số vượt vượt quá 6000
+ reposts_count: Số lượt repost khá ít dưới 300 lượt và có 1 số track 
+ download_count: Số lượt download trong khoảng từ 0-5000 lượt và có 2 track được 16000 và 26000
+ comment_count: Đa số lượt comment trong khoảng 0-180 và có 1 số vượt quá 200
+ playback_count: Lượt playback nằm trong khoảng 0-100000 và có nhưng playback vượt qua 200000

### 3. Phân tích dữ liệu đa biến

```python
track = new_track.select_dtypes(include = ['int64'])
d = track.corr(method='pearson')
f, ax = plt.subplots(figsize=(10,10))
mask = np.triu(np.ones_like(track.corr()))
sns.heatmap(d, ax=ax, cmap="YlGnBu", annot=True, square=True, mask=mask)
plt.show()
```

#### Nhận xét:
+ Dựa vào biểu đồ trên ta thấy sự tương quan của các dữ liệu sau:
    + reposts_count và likes_count: track được yêu thích càng nhiều thì càng được tài khoản khác chia sẻ nhiều hơn và ngược lại.
    + reposts_count và playback_count: track được nghe càng nhiều thì càng đuợc tài khoản khác chia sẻ nhiều hơn và ngược lại.
    + likes_count và comments_count: track càng được yêu thích thì càng nhiều người bình luận và ngược lại.
    + playback_count và comments_count: track được nghe càng nhiều thì càng nhiều người bình luận và ngược lại.
    + full_duration và playback_count không ảnh hưởng gì đến nhau

```python
plt.figure(figsize=(8, 8))
plt.scatter(new_track["likes_count"], new_track["reposts_count"], c="green")
plt.title("Biểu đồ phân tán giữa likes_count và reposts_count")
plt.xlabel("likes_count")
plt.ylabel("reposts_count")
plt.show()
```

Ta thấy rõ ràng về mối quan hệ đồng biến giữa 2 thuộc tính likes và reposts

```python
plt.figure(dpi=120)
plt.scatter(new_track["likes_count"], new_track["comment_count"], s=new_track["reposts_count"] / 50, alpha=0.5)
plt.xlabel("likes_count")
plt.ylabel("comment_count")
plt.show()
```

Nhận xét: lượt like và comment có ảnh hưởng mật thiết đến lượt download của bài hát. Cụ thể thì like và comment tăng thì lượt download sẽ tăng và ngược lại
