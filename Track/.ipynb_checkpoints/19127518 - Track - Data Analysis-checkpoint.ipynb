### Import Thư Viện

```python
import pandas as pd
import numpy as np
import requests as r
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

### 1. Đọc dữ liệu từ file CSV đã được xử lí

```python
new_track = pd.read_csv('New_Track.csv')
new_track
```

```python
new_track.info()
```

#### Tóm tắt thông tin về 9 thuộc tính của tập dữ liệu Track:

|Thuộc tính|Kiểu dữ liệu|     Ý nghĩa  
|:--------:|:-------------:|:-----------:
|**id**|Số nguyên|Định danh cho mỗi track|
|**full_duration**|Số nguyên|Thời gian của từng track|   
|**genre**|Chuỗi|Thể loại nhạc của mỗi track| 
|**comment_count**|Số nguyên|Số lượt bình luận track đó|
|**likes_count**|Số nguyên|Số lượt thích track đó|
|**playback_count**|Số nguyên|Số lượt nghe lại track đó|   
|**download_count**|Số nguyên|Số lượt tải về của track đó| 
|**display_date**|Chuỗi|Thời gian xuất bản track lên Soundcloud|

### 2. Phân tích dữ liệu đơn biến

```python
track_num = new_track.select_dtypes(include = ['float64', 'int64'])
track_num.describe().T                                
```

#### Biểu đồ tuần suát của các biến numeric

```python
track_num.hist(figsize=(16, 16), bins=150, xlabelsize=8, ylabelsize=8);
```

#### Nhận xét sơ bộ: 
    + full_duration: Các track đa số kéo dài khoảng 5 - 10 phút
    + comment_count, reposts_count và download_count: Rất ít đa số track đều rất ít có comment, repost và download.
    + likes_count: Có só lượt yêu thích trong khoảng từ 0 đến 2000 
    + playback_count: Có số lượt playback trong khoảng từ 0 đến 50000

### 3. Phân tích dữ liệu đa biến

```python
correlation_matrix = new_track.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix[(correlation_matrix >= 0.5) | (correlation_matrix <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

#### Nhận xét:
+ Dựa vào biểu đồ trên ta thấy sự tương quan của các dữ liệu sau:
    + reposts_count và likes_count: track được yêu thích càng nhiều thì càng được tài khoản khác chia sẻ nhiều hơn và ngược lại.
    + reposts_count và playback_count: track được nghe càng nhiều thì càng đuợc tài khoản khác chia sẻ nhiều hơn và ngược lại.
    + likes_count và comments_count: track càng được yêu thích thì càng nhiều người bình luận và ngược lại.
    + playback_count và comments_count: track được nghe càng nhiều thì càng nhiều người bình luận và ngược lại.

### 4. Phân tích dữ liệu rời rạc

```python
plt.figure(figsize=(50, 50))
sns.countplot(y = "genre" , data = new_track)
plt.title('Số lượng playlist được tìm kiếm nhiều nhất theo thể loại nhạc');
```
